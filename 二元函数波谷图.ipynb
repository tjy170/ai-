```python
import numpy as np
import matplotlib.pyplot as plt

# 定义具有两个波谷的一元函数
def f(x):
    return np.exp(-(x-2)**2) + np.exp(-(x+2)**2)

# 计算函数的梯度
def df(x):
    return -2 * (x-2) * np.exp(-(x-2)**2) - 2 * (x+2) * np.exp(-(x+2)**2)

# 梯度下降算法
def gradient_descent(x0, learning_rate, num_iterations):
    x = x0
    for i in range(num_iterations):
        grad = df(x)
        x = x - learning_rate * grad
    return x

# 参数设置
x0 = np.random.uniform(-5, 5)  # 随机初始化
learning_rate = 0.1  # 学习率
num_iterations = 100  # 迭代次数

# 执行梯度下降算法
x_min = gradient_descent(x0, learning_rate, num_iterations)

# 绘制函数图像和梯度下降路径
x_values = np.linspace(-5, 5, 400)
y_values = f(x_values)
plt.plot(x_values, y_values, label='f(x)')

# 绘制梯度下降路径
plt.scatter([x_min], [f(x_min)], color='red', label='Minimum (approx)')

plt.legend()
plt.show()

print(f"Approximate minimum at x = {x_min} with f(x) = {f(x_min)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义具有两个波谷的二元函数
def f(x, y):
    return np.exp(-(x-2)**2 - (y-2)**2) + np.exp(-(x+2)**2 - (y+2)**2)

# 计算函数的梯度
def df(x, y):
    dx = -2 * (x-2) * np.exp(-(x-2)**2 - (y-2)**2) - 2 * (x+2) * np.exp(-(x+2)**2 - (y+2)**2)
    dy = -2 * (y-2) * np.exp(-(x-2)**2 - (y-2)**2) - 2 * (y+2) * np.exp(-(x+2)**2 - (y+2)**2)
    return np.array([dx, dy])

# 梯度下降算法
def gradient_descent(x0, y0, learning_rate, num_iterations):
    x, y = x0, y0
    for i in range(num_iterations):
        grad = df(x, y)
        x, y = x - learning_rate * grad[0], y - learning_rate * grad[1]
    return x, y

# 参数设置
x0, y0 = np.random.uniform(-5, 5, 2)  # 随机初始化
learning_rate = 0.1  # 学习率
num_iterations = 1000  # 迭代次数

# 执行梯度下降算法
x_min, y_min = gradient_descent(x0, y0, learning_rate, num_iterations)

# 绘制函数图像
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# 绘制梯度下降路径
ax.scatter([x_min], [y_min], [f(x_min, y_min)], color='red', label='Minimum (approx)')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()
plt.show()

print(f"Approximate minimum at x = {x_min}, y = {y_min} with f(x, y) = {f(x_min, y_min)}")
```
